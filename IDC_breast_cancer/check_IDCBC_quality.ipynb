```python
import numpy as np
import pandas as pd
import anndata as ad
import matplotlib.pyplot as plt
```

```python
# Load Data & Run
fileAlpha = '/shared/space2/molina/suttyg/FASTQ_Cancer/fouriercycle/scrna_2p0_1p5.h5ad'
adata = ad.read_h5ad(fileAlpha)
targetnames = adata.obs.index
theta_smooth = np.round(np.linspace(0.01, 1.00, 100), 2)  # 100 bins from 0.01 to 1.00
alpha = pd.DataFrame(adata.obsm["alpha"]).to_numpy()
```

```python
#CCNE1, CCNE2, NUSAP1, AURKA, ARUKB, CCNB1, CDK1
gene_nb = np.where(targetnames == 'CDK1')[0][0]

plt.figure(figsize=(6, 4))
plt.title(targetnames[gene_nb])
plt.xlabel("\u03B8")
plt.ylabel("Transcription rate")
#plt.ylim(-0.75, 0.75)
plt.plot(theta_smooth,alpha[gene_nb,:], label="scRNA-seq", color="green")
plt.show()
```
